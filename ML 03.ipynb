```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=dataset)
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
min_range =dataset["CoapplicantIncome"].mean()-3*dataset["CoapplicantIncome"].std()
max_range=dataset["CoapplicantIncome"].mean()+3*dataset["CoapplicantIncome"].std()
```

```python
min_range,max_range
```

```python
new_data=dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_data)
```

```python
z_score=(dataset["CoapplicantIncome"]-dataset["CoapplicantIncome"].mean())/(dataset["CoapplicantIncome"].std())
```

```python
dataset["z_score"]= z_score
```

```python
dataset[dataset["z_score"]<3]
```

# Scaling

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome"])

plt.show()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])

plt.show()
```
