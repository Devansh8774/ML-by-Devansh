```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv("social network ads.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=["EstimatedSalary"],inplace =True)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
x= dataset[["Age"]]
y= dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr.predict(x),data=dataset ,color='red')

plt.show()
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("placement.csv")
```

```python
dataset.head(3)
```

```python
# plt.figure(figsize=(3,4))
sns.scatterplot(x="cgpa",y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python
x= dataset.iloc[:,:-1]
y =dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[8.14 ,6.52]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()                      
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
```

```python
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train , y_test = train_test_split(x, y ,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
lr.score(x_test , y_test)*100
```
