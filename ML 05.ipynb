```python
import pandas as pd 
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv("diabetes.csv")
```

```python
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=3,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

# Supervised machine learning(Regression Analysis)

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("placement.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns. scatterplot(x="cgpa",y="iq",data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["cgpa"]]
y=dataset["iq"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.8]])
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="iq",data=dataset)
plt.plot(dataset["cgpa"],y_prd,c="red")
plt.legend(["org data","predict line"])
plt.savefig("predict.jpg")
plt.show()
```
